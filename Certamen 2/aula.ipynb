```python
import numpy as np
import scipy.linalg as sp

def test_matrix(n,a,b,c):
    aux = np.zeros((n,n))
    for i in range(0,n):
        aux[i,n-i-1] = a
        if ((n-i)) < n:
            aux[i,(n-i)] = c
        if ((n-i)-2) >= 0:
            aux[i,(n-i)-2] = b
    return aux

    

def test_b(n):
    a = np.arange(1,n+1)
    return 1/a
    


def permutations(n):
    permutation_m = np.identity(n)
    aux = np.copy(permutation_m)
    for i in range(1,n//2+1):
        up = aux[i-1]
        down = aux[n-i]
        permutation_m[i-1] = down
        permutation_m[n-i] = up
    return  permutation_m


def gauss_seidel_solver(A, b, n_iter=50):
    n = A.shape[0]
    X = np.empty((n_iter, n))
    X[0] = np.zeros(n)
    R = np.tril(A) 
    U = A-R
    for i in range(1, n_iter):
       
        X[i] = X[i-1]+np.linalg.solve(R, b-np.dot(A, X[i-1]))
    return X[-1]

def solver(a,b):
    dimensiones = a.shape
    aux = np.copy(a)
    for i in range(0,dimensiones[0]):
        if i == 0:
            aux[i,i] = aux[i,i]
        else:
            coeficiente = aux[i,i-1]/aux[i-1,i-1]
            aux[i,i] = aux[i,i] - (aux[i-1,i]*coeficiente)
            aux[i,i-1] = aux[i,i-1] - (aux[i-1,i-1]*coeficiente)
            b[i] = b[i] - (b[i-1]*coeficiente) 
    X = sp.solve_triangular(aux, b)
    
    return X
    


def vector_sol(n,a,b,c)  :
    A = test_matrix(n,a,b,c)
    P = permutations(n)
    B = test_b(n)
    solv = solver(P.dot(A),P.dot(B))
    return solv





```

```python
A = test_matrix(3000,10,-4,0.5)
P = permutations(3000)
B = test_b(3000)

test_sol = vector_sol(3000,10,-4,0.5)

print(test_sol)
```

# Pregunta 3:

```python
def bisymmetric(n):
    np.random.seed(123)
    J = np.eye(n)[::-1]
    A = np.random.random((n, n))
    A = np.dot(A.T, A) / n ** 2
    A = np.dot(A, J) + np.dot(J, A)
    return A
```

```python
def bisymmetricZ(n):
    A = bisymmetric(n)
    A = np.max(np.abs(np.diag(A))) * np.eye(n) - A
    return A
```

```python
def bisymmetricZDD(n, delta):
    A = bisymmetricZ(n)
    A -= np.diag(np.diag(A))
    C = np.diag(np.abs(np.sum(A, axis=1)))
    A += C + delta * np.eye(n)
    return A
```

```python
def generateSystem(n, delta=None):
    if delta == None: # Generacion de A para la pregunta (b)
        A = bisymmetricZ(n)
    else:
        A = bisymmetricZDD(n, delta) # Generacion de A para la pregunta (d) y (e)
    b = np.dot(A, np.ones(n))
    return A, b # Sistema de ecuaciones lineales
```
